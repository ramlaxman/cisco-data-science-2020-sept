```python
# (3) Choose an estimator (K Network Neighbors)

# assume we have data
# - there are 2 classes
# - there are 2 features

# ask 1 neighbor: green triangle
# 3 neighbors: blue circle
# 5 neighbors: blue circle
# 7 neighbors: green triangle
```

```python
from sklearn.datasets import load_iris
from sklearn.neighbors import KNeighborsClassifier

iris = load_iris()

X = DataFrame(iris.data, columns=iris.feature_names)
y = Series(iris.target)

model = KNeighborsClassifier()  # create a new, "naked" model -- brains but no experience

model.fit(X, y)  # give the model experience -- show it flower data + classification

model.predict([ [1,2,3,4],   # versicolor
                [5,6,7,8] ])    # virginica
```

```python
y_pred = model.predict(X)   # make predictions based on X, and put into y_pred
```

```python
(y_pred == y).value_counts()
```

```python
145 / 150
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_pred, y)
```

```python
# overfitting is THE BIGGEST PROBLEM in modeling
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
accuracy_score(y_pred, y_test)
```

```python
from sklearn.datasets import load_iris
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

iris = load_iris()

X = DataFrame(iris.data, columns=iris.feature_names)
y = Series(iris.target)

model = KNeighborsClassifier()  

all_scores = []
for i in range(1000):
    X_train, X_test, y_train, y_test = train_test_split(X, y)
    model.fit(X_train, y_train)

    y_pred = model.predict(X_test)
    all_scores.append(accuracy_score(y_pred, y_test))
    
Series(all_scores).describe()
```
